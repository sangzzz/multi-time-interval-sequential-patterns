```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
```

```python
alpha = 11
label_encoder = LabelEncoder()
```

```python
unprepared_dataset = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/DWDM - CS416/Jewelry_unprepared.csv')
```

```python
unprepared_dataset.sample(10)
```

```python
unprepared_dataset.nunique()
```

```python
unprepared_dataset.shape
```

```python
def formatEventDate(dateString):
  dateString = str(dateString)[:-4]
  dt = datetime.strptime(dateString, '%Y-%m-%d %H:%M:%S')
  day_tag = 1 + dt.day // alpha
  return str(dt.year) + "_" + str(dt.month) + "_" + str(day_tag)
formatEventDate("2018-11-29 16:30:45 UTC")
```

```python
def getDay(dateString):
  dateString = str(dateString)[:-4]
  dt = datetime.strptime(dateString, '%Y-%m-%d %H:%M:%S')
  return str(dt.day)
```

```python
unprepared_dataset['timestamp'] = unprepared_dataset.event_time.apply(formatEventDate)
```

```python
unprepared_dataset['day'] = unprepared_dataset.event_time.apply(getDay)
```

```python
unprepared_dataset['item_name'] = unprepared_dataset['category_id'].astype(str) + '_' + unprepared_dataset['category_code'].astype(str) + '_' + unprepared_dataset['metal'].astype(str) + '_' + unprepared_dataset['color'].astype(str) + '_' 
```

```python
unprepared_dataset.sample(10)
```

```python
unprepared_dataset.loc[unprepared_dataset['quantity']!=1, ['quantity']]
```

```python
unprepared_dataset.isnull().sum()
```

```python
unprepared_dataset.drop(unprepared_dataset[unprepared_dataset['quantity'] != 1].index, inplace=True)
unprepared_dataset.drop(labels=['event_time', 'order_id', 'brand', 'price', 'user_id', 'gender', 'gem', 'product_id', 'quantity'], axis=1, inplace=True)
unprepared_dataset.dropna(axis=0, how='any', inplace=True)
```

```python
unprepared_dataset.sample(10)
```

```python
print('Count of null Values:\n', unprepared_dataset.isnull().sum(), end='\n\n')
print('Shape of Dataset: ', unprepared_dataset.shape, end='\n\n')
print('Number of Unique values in each series:\n', unprepared_dataset.nunique(), end='\n')
```

```python
unprepared_dataset['item_name'] = label_encoder.fit_transform(unprepared_dataset['item_name'])
np.save('item_classes.npy', label_encoder.classes_)
```

```python
unprepared_dataset['transaction_id'] = label_encoder.fit_transform(unprepared_dataset['timestamp'])
np.save('transaction_classes.npy', label_encoder.classes_)
```

```python
unprepared_dataset['item'] = tuple(zip(unprepared_dataset['item_name'], unprepared_dataset['day']))
unprepared_dataset.drop(['day', 'item_name', 'category_code', 'category_id', 'color', 'metal', 'timestamp'], inplace=True, axis=1)
unprepared_dataset.sample(10)
```

```python
prepared_dataset = unprepared_dataset
```
